```

%%writefile app.py

import streamlit as st
import numpy as np
from PIL import Image
import pickle as pk
import tensorflow as tf
from keras.models import Sequential
from keras.models import load_model
import os
os.chdir('/content/drive/My Drive/Projects/untitled')
from dump import Execute_DB

 
def predict_catagory(test_model, data):
    import os 
    from dump import Test

    output = Test.Predict(data)  
    
    return output

def mapping_data(val):
    if (val.lower() == 'yes') or (val.lower() == 'male') :
        return 1
    elif (val.lower() == 'no') or (val.lower() == 'female'):
        return 0
    else:
        return 0
                
def main():
    st.set_page_config(layout="wide")
    test_model = r'/content/drive/My Drive/Projects/untitled/Model_file.pkl'
    
    #models = [classifier_logistic, classifier_naive, classifier_rf]    
    
    st.title('''
             My Pocket Tracker For COVID-19
             ''')
    
    st.text('''
            We looked at easy to build open-source techniques leveraging AI 
            which can give us state-of-the-art accuracy in detecting the Novel 
            COVID-19 virus thus enabling AI for social good.

            ''')  
    
    try:
        name = st.text_input("Please Enter your Name : ")
        fever = st.radio('Is there Any kind of Fever that you experiece in your body ? ',('Yes','No'))
        tiredness = st.radio('Is there any kind of Tiredness in your body ? ',('Yes','No'))
        dry_cough = st.radio('Do You Feel Something like Dry Cough ? ',('Yes','No'))
        breathing = st.radio('Any Feelings Regarding Difficulty-in-Breathing ? ',('Yes','No'))
        throat = st.radio('How do you feel inside Throat, is it Sore or not ? ',('Yes','No'))
        nasal = st.radio('Is there any kind of Nasal-Congestion that you feels in nose ? ',('Yes','No'))
        nose = st.radio('What about condition of your nose, Do you have a Runny Nose ? ',('Yes','No'))
        diarrhea = st.radio('Do you feel any Diarrhea or loose motions related issue ? ',('Yes','No'))
        gender = st.radio('Who are you, Male or Female ? ',('Male','Female','Others'))
        contact = st.radio('Do you have any Contact with any Covid Positive patient ? ',('Yes','No','Dont Know'))
        age = st.text_input("What's Your Age : ")
            
        data = [fever, tiredness, dry_cough, breathing, throat, nasal, nose, diarrhea, gender, contact, age]

        for_sql = (name, fever, age, tiredness, dry_cough, breathing, throat, nasal, nose, diarrhea, gender, contact)
        data = list(map(mapping_data,data))



        if st.button('Make a Test'):
            output = predict_catagory(test_model, data)
            st.success("We Feel That your Symptoms says &nbsp : &nbsp {} infection of COVID :sunglasses:".format(output)) 
            Execute_DB(for_sql+(output,))           
        
    
    except AttributeError:
        pass

if __name__=='__main__':
    main()
```
